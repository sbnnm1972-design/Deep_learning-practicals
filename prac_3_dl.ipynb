```python


import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import DataLoader
from torchtext.datasets import IMDB
from torchtext.data.utils import get_tokenizer
from torchtext.vocab import build_vocab_from_iterator
```

```python
from datasets import load_dataset

dataset = load_dataset("imdb")

train_ds = dataset["train"]
test_ds = dataset["test"]

print(train_ds[0])
```

```python
tarin_it,test_it = IMDB(split=('train','test'))
```

```python
tokenizer = get_tokenizer('basic_english')
def yield_tokens(data_iter):
  for _,text in data_iter:
    yield tokenizer(text)
vocab= build_vocab_from_iterator(yield_tokens(train_ds),specials=["<pad>","<unk>"])
vocab.set_default_index(vocab["<unk>"])
```

```python
def encode(text):
  tokens = tokenizer(text)
  return vocab(tokens)

def encode_lab(label):
  return 1 if label == "pos" else 0

def collate_batch(batch):
  texts =[]
  label_list= []

  for labels,text in batch:
    texts.append(torch.tensor(encode(text),dtype=torch.long))
    label_list.append(torch.tensor(encode_lab(labels)))
    texts =torch.nn.utils.rnn.pad_sequence(texts,batch_first=True,padding_value=vocab["<pad>"])
    labels = torch.tensor(label_list,dtype=torch.float32)
    return texts,labels
```

```python
train_loader = DataLoader(train_ds,
                          batch_size=32,shuffle=True,collate_fn=collate_batch)

test_loader = DataLoader(test_ds,
                         batch_size=32,shuffle=False,collate_fn=collate_batch)
```

```python
class IMDBclassifier(nn.Module):
  def __init__(self,vocab_size,embd_dim):
    super().__init__()

    self.embedding = nn.Embedding(vocab_size,embd_dim)
    self.fc = nn.Linear(embd_dim,1)

  def forward(self,x):
    x = self.embedding(x)
    x = x.mean(dim=1)
    logits = self.fc(x)
    return logits
```

```python
def train_nn(model , train_loader , epochs=5):
  model.train()

  for  epoch in range(epochs):
    total_loss=0
    correct = 0
    total =0

    for X_batch , y_batch in train_loader:
      optimizer.zero_grad()
      logits =model(X_batch).squeeze(1)
      loss = criterion(logits, y_batch)
      loss.backward()
      optimizer.step()

      preds = (torch.sigmoid(logits) > 0.5).long()
      correct += (preds == y_batch).sum().item()
      total += y_batch.size(0)
      total_loss += loss.item()

    acc = 100*correct/total
    print(f"Epoch {epoch+1}| Loss: {total_loss:.4f}| acc {acc:.2f}%")
```

```python
def eval(model,test_loader):
  model.eval()
  correct = 0
  total = 0

  with torch.no_grad():
    for X_batch , y_batch in test_loader:
      logits = model(X_batch).squeeze(1)
      preds = (torch.sigmoid(logits)> 0.5).long()
      correct += (preds == y_batch).sum().item()
      total += y_batch.size(0)

  print(f"test acc:{100*correct/total:.2f}%")
```

```python
model= IMDBclassifier(vocab_size= 10000,embd_dim=100)
```

```python
criterion = nn.BCEWithLogitsLoss()
optimizer = optim.Adam(model.parameters(),lr=0.001)
```

```python
train_nn(model,train_loader,epochs=5)
```

```python
eval(model,test_loader)
```
