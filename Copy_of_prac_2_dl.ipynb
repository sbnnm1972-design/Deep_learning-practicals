```python
import numpy as  np
import torch
from torch import nn
from torch.nn import functional as F
from torch.utils.data import DataLoader,TensorDataset
from sklearn.model_selection import train_test_split
from sklearn.datasets import make_classification
```

```python
X, Y = make_classification(n_samples=500 , n_features=6 ,n_classes= 2, random_state = 42)

X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.2,random_state=42)

X_tr_tensor = torch.tensor(X_train,dtype=torch.float32)
Y_tr_tensor = torch.tensor(Y_train,dtype=torch.long)
X_ts_tensor = torch.tensor(Y_test,dtype=torch.float32)
Y_ts_tensor = torch.tensor(X_test,dtype=torch.long)

train_data = TensorDataset(X_tr_tensor,Y_tr_tensor)
train_load = DataLoader(train_data,batch_size=25,shuffle=True)
test_data= TensorDataset(X_ts_tensor,Y_ts_tensor)
test_load = DataLoader(test_data,batch_size=25,shuffle=True)

class singlelayer(nn.Module):
  def __init__(self,input_dim, hidden_dim , n_class) :
    super().__init__()

    self.fc1 = nn.Linear(input_dim,hidden_dim)
    self.fc2 = nn.Linear(hidden_dim,n_class)

  def forward(self,x):
    x = self.fc1(x)
    x= F.relu(x)
    x = self.fc2(x)
    return x
```

```python

model = singlelayer(input_dim=6, hidden_dim=16, n_class=2)
```

```python
def train_nn(model , train_load , epochs=100,lr =0.01):
  criterion= nn.CrossEntropyLoss()
  optimizer = torch.optim.SGD(model.parameters(),lr=lr)
  for epoch in range(epochs):
    correct =  0
    total = 0
    running_loss= 0.0
    for X_batch, y_batch in train_load:
            optimizer.zero_grad()

            logits = model(X_batch)
            loss = criterion(logits, y_batch)

            loss.backward()
            optimizer.step()

            running_loss += loss.item()

            preds = torch.argmax(logits, dim=1)
            correct += (preds == y_batch).sum().item()
            total += y_batch.size(0)
    if epoch% 10 == 0:
      pred = torch.argmax(logits , dim=1)
      acc = 100*correct/total
      print(f"epoch{epoch:03d} | loss: {running_loss:.4f} | acc:{acc:.2f}%")
```

```python
train_nn(model,train_load)
```
